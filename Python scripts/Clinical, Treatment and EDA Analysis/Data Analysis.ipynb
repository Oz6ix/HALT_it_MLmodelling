```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Convert categorical variables to numeric
categorical_cols = [
    'sex', 'location', 'haematemesis', 'melaena', 'variceal', 
    'shock', 'bleeding', 'comorbcv', 'comorbresp', 'comorbliver',
    'comorbrenal', 'comorbmalig', 'comorbother', 'anticoag', 
    'emergency', 'diagnosticendoscopic', 'therapeuticendoscopic',
    'diagnosticradiological', 'therapeuticradiological', 'surgicalintervention',
    'causebleed', 'doseone', 'dosetwo', 'helicobacter', 'h2', 'proton', 
    'vasopressin', 'antibiotics', 'antifibrinolytics', 'transfused'
]

# Convert categorical variables using one-hot encoding
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)

# List the columns to identify the new names
print("Columns after encoding:", data.columns)

# Manually identify the columns corresponding to the symptoms of interest
# For example:
symptoms = [
    'haematemesis_Yes', 'melaena_Yes', 'shock_Yes', 'bleeding_Yes',
    'variceal_Yes'
]

# Ensure all required columns are present
for symptom in symptoms:
    if symptom not in data.columns:
        print(f"Warning: {symptom} not found in the DataFrame.")

# Perform logistic regression for each outcome
outcomes = ['rebleeding', 'mortality']
for outcome in outcomes:
    print(f"Logistic Regression for {outcome}")
    
    # Add constant for intercept
    X = sm.add_constant(data[symptoms])
    
    # Define the dependent variable (outcome)
    y = data[outcome]
    
    # Fit the logistic regression model
    model = sm.Logit(y, X).fit()
    
    # Print the odds ratios
    odds_ratios = np.exp(model.params)
    print(odds_ratios)
    
    # Print the summary of the model
    print(model.summary())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Mapping the old names to new readable names
symptom_mapping = {
    'haematemesis_Yes': 'Hematemesis',
    'melaena_Yes': 'Melaena',
    'shock_Yes': 'Shock',
    'variceal_Yes': 'Variceal'
}

odds_ratios_rebleeding = pd.Series({
    'haematemesis_Yes': np.exp(-0.1829),
    'melaena_Yes': np.exp(0.1391),
    'shock_Yes': np.exp(0.4452),
    'variceal_Yes': np.exp(-0.3984)
})

odds_ratios_mortality = pd.Series({
    'haematemesis_Yes': np.exp(0.0085),
    'melaena_Yes': np.exp(-0.2749),
    'shock_Yes': np.exp(0.8235),
    'variceal_Yes': np.exp(0.1689)
})

# Rename the indices using the symptom mapping
odds_ratios_rebleeding.index = odds_ratios_rebleeding.index.map(symptom_mapping)
odds_ratios_mortality.index = odds_ratios_mortality.index.map(symptom_mapping)

# Combine the two series into a DataFrame for easier plotting
odds_ratios_combined = pd.DataFrame({
    'Rebleeding': odds_ratios_rebleeding,
    'Mortality': odds_ratios_mortality
})

# Plotting the combined odds ratios
plt.figure(figsize=(10, 6))
odds_ratios_combined.plot(kind='barh')
plt.title('Odds Ratios for Rebleeding and Mortality')
plt.xlabel('Odds Ratio')
plt.ylabel('Symptoms')
plt.axvline(x=1, color='red', linestyle='--')  # Reference line for odds ratio of 1
plt.legend(title='Outcome')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
from scipy.stats import chi2_contingency

# Function to calculate and print chi-square test results
def chi_square_test(data, feature, target):
    contingency_table = pd.crosstab(data[feature], data[target])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Chi-square test for {feature} and {target}:")
    print(f"Chi2: {chi2}, p-value: {p}\n")

# Variables to test
clinical_vars = ['sex', 'location', 'haematemesis', 'melaena', 'variceal', 'shock', 'bleeding']

# Test relationship between clinical variables and rebleeding
for var in clinical_vars:
    chi_square_test(data, var, 'rebleeding')

# Visualize the relationship
for var in clinical_vars:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=var, hue='rebleeding', data=data, palette='Set1')
    plt.title(f'Relationship between {var} and Rebleeding')
    plt.show()
```

Based on the chi-square test results, we can draw some conclusions regarding the relationships between clinical presentations and rebleeding:

	1.	Sex and Rebleeding: There is no significant relationship between sex and rebleeding (p-value: 0.1129).
	2.	Location and Rebleeding: There is a significant relationship between the location of bleeding and rebleeding (p-value: 1.50e-09). Upper gastrointestinal bleeding is more common, and it is associated with a higher rate of rebleeding compared to lower gastrointestinal bleeding.
	3.	Haematemesis and Rebleeding: There is a significant relationship between haematemesis and rebleeding (p-value: 2.89e-05). Patients with haematemesis are more likely to experience rebleeding.
	4.	Melaena and Rebleeding: There is a significant relationship between melaena and rebleeding (p-value: 0.0088). Patients with melaena are more likely to experience rebleeding.
	5.	Variceal and Rebleeding: There is a significant relationship between variceal bleeding and rebleeding (p-value: 1.84e-08). Patients with variceal bleeding are more likely to experience rebleeding.
	6.	Shock and Rebleeding: There is a significant relationship between shock and rebleeding (p-value: 9.31e-09). Patients who present with shock are more likely to experience rebleeding.
	7.	Bleeding and Rebleeding: There is no significant relationship between bleeding and rebleeding (p-value: 0.1119).

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, mannwhitneyu

# Load the dataset
df = pd.read_csv('cleaned_data.csv')  # Replace with your actual dataset path

# Function to calculate p-value for categorical data
def calculate_p_value_categorical(data, column, outcome):
    contingency_table = pd.crosstab(data[column], data[outcome])
    _, p_value, _, _ = chi2_contingency(contingency_table)
    return p_value

# Function to calculate p-value for continuous data
def calculate_p_value_continuous(data, column, outcome):
    group_no = data[data[outcome] == 0][column]
    group_yes = data[data[outcome] == 1][column]
    _, p_value = mannwhitneyu(group_no, group_yes)
    return p_value

# Function to get summary statistics for categorical data
def get_categorical_summary(data, column, outcome):
    summary = data.groupby(outcome)[column].value_counts(normalize=False).unstack().fillna(0)
    percentages = data.groupby(outcome)[column].value_counts(normalize=True).unstack().fillna(0) * 100
    summary = summary.astype(int).astype(str) + ' (' + percentages.round(1).astype(str) + '%)'
    p_value = calculate_p_value_categorical(data, column, outcome)
    return summary, p_value

# Function to get summary statistics for continuous data
def get_continuous_summary(data, column, outcome):
    summary = data.groupby(outcome)[column].apply(lambda x: f"{x.median()} ({x.quantile(0.25)}-{x.quantile(0.75)})")
    p_value = calculate_p_value_continuous(data, column, outcome)
    return summary, p_value

# List of characteristics to include in the table
characteristics = [
    ('age', 'continuous'),
    ('sex', 'categorical'),
    ('comorbcv', 'categorical'),
    ('comorbresp', 'categorical'),
    ('comorbliver', 'categorical'),
    ('comorbrenal', 'categorical'),
    ('comorbmalig', 'categorical'),
    ('comorbother', 'categorical'),
    ('haematemesis', 'categorical'),
    ('melaena', 'categorical'),
    ('emergency', 'categorical'),
    ('variceal', 'categorical'),
    ('anticoag', 'categorical'),
    ('sbp', 'continuous'),
    ('heartrate', 'continuous'),
    ('shock', 'categorical'),
    ('diagnosticendoscopic', 'categorical'),
    ('diagnosticradiological', 'categorical'),
    ('helicobacter', 'categorical'),
    ('h2', 'categorical'),
    ('proton', 'categorical'),
    ('vasopressin', 'categorical'),
    ('transfused', 'categorical')
]

# Outcomes to include in the table
outcomes = ['rebleeding', 'mortality', 'long_admission', 'icu_admission', 'hdu_admission']

# Generate the table
tables = {}
for outcome in outcomes:
    rows = []
    n_no = df[df[outcome] == 0].shape[0]
    n_yes = df[df[outcome] == 1].shape[0]
    for characteristic, char_type in characteristics:
        if char_type == 'categorical':
            summary, p_value = get_categorical_summary(df, characteristic, outcome)
        else:
            summary, p_value = get_continuous_summary(df, characteristic, outcome)
        
        # Split male and female for sex characteristic
        if characteristic == 'sex':
            row_male = {
                'Characteristic': 'Male',
                'No (n={})'.format(n_no): summary.loc[0, 'Male'] if 0 in summary.index and 'Male' in summary.columns else 'N/A',
                'Yes (n={})'.format(n_yes): summary.loc[1, 'Male'] if 1 in summary.index and 'Male' in summary.columns else 'N/A',
                'p-Value': p_value
            }
            row_female = {
                'Characteristic': 'Female',
                'No (n={})'.format(n_no): summary.loc[0, 'Female'] if 0 in summary.index and 'Female' in summary.columns else 'N/A',
                'Yes (n={})'.format(n_yes): summary.loc[1, 'Female'] if 1 in summary.index and 'Female' in summary.columns else 'N/A',
                'p-Value': p_value
            }
            rows.extend([row_male, row_female])
        else:
            row = {
                'Characteristic': characteristic,
                'No (n={})'.format(n_no): summary.loc[0] if 0 in summary.index else 'N/A',
                'Yes (n={})'.format(n_yes): summary.loc[1] if 1 in summary.index else 'N/A',
                'p-Value': p_value
            }
            rows.append(row)
    
    # Adding sub-headings and restructuring the table
    table = pd.DataFrame(rows)
    formatted_table = pd.DataFrame(columns=['Characteristic', 'No (n=...)', 'Yes (n=...)', 'p-Value'])
    
    sections = {
        'Demographics': ['Age', 'Male', 'Female'],
        'Comorbidities': ['comorbcv', 'comorbresp', 'comorbliver', 'comorbrenal', 'comorbmalig', 'comorbother'],
        'Associated Symptoms and Signs': ['haematemesis', 'melaena', 'emergency', 'variceal', 'anticoag'],
        'Vital Signs': ['sbp', 'heartrate', 'shock'],
        'Diagnostic': ['diagnosticendoscopic', 'diagnosticradiological', 'helicobacter'],
        'Treatment': ['h2', 'proton', 'vasopressin', 'transfused']
    }
    
    for section, chars in sections.items():
        formatted_table = pd.concat([formatted_table, pd.DataFrame([{'Characteristic': f"**{section}**"}]), table[table['Characteristic'].isin(chars)]])
    
    tables[outcome] = formatted_table.reset_index(drop=True)

# Display the tables
for outcome, table in tables.items():
    print(f"### {outcome.capitalize()} ###")
    print(table.to_string(index=False))
    print("\n")

# Optionally, save tables to CSV
for outcome, table in tables.items():
    table.to_csv(f'{outcome}_table.csv', index=False)
```
